Ejercicio 1.- Funciones en matrices
 
  1.Suma.- sean dos matrices $A$ y $B$ de dimensión $n$, entonces existe una matiz $C$ t.q. $$A + B = C$$ donde $C$ es de dimensión $n$.
  
  1.1.Asociatividad
       $$A + (B+C) = (A+B)+C$$
  1.2.Conmutatividad
    $$A+B=B+A$$
  1.3.Nuetro aditivo
    Sea $C$ la matriz llena de ceros y de dimensión $n$, entonces: $$A+C=A$$
  
  
  2.Multiplicación por un escalar o una matriz
  
  2.1.Matriz por un escalar, sea $k$ que pertenece a $R$ y $a_{ij}$ las entradas de $A$: $$kA=k(a_{ij})$$
  
  2.2.Multiplicación de matrices, sea $A$ una matriz de $nxm$ y $B$ una matriz de $mxn$, entonces:
    
   $AxB = C$ donde $C$ es una matriz de $nxn$
     
   $AxB = D$ donde $D$ es una matriz de $mxm$
      
    La multiplicación no es conmutativa.
    
  2.3.Asociatividad
       $$A(BC) = (AB)C$$
  2.4.Neutro Multiplicativo, sea $I$ la matriz identidad t.q. $$AxI=A$$ Existe una matriz B t.q. $$AxB=I$$ a $B$ la llamaremos la matriz inversa de $A$.
