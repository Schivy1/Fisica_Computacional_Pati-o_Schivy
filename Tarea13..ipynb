#### Respuesta 
Realizando la sustitución tenemos un pliniomio muy sencillo para encontrar sus valores propios y como esta igualada a cero no tiene solución particular, por lo tanto encontramos que:
$$m \ddot{x} + b \dot{x} + kx = 0$$
Resolviendo el polinomio caracteristico tenemos que:
$$x(t) = \frac{2 A cos(\frac{(3t)^\frac{1}{2}}{2})}{e^\frac{t}{2}}$$

```julia
using PyPlot
```

```julia
function Runge(f::Function,x0,y0,t0,h,N) 
    T = []
    X = []
    Y = []
    t = t0
    x = x0
    y = y0
    while t<=N
        push!(T,t)
        push!(X,x)
        push!(Y,y)
        k1=(f(t,x)) 
        k2=(f(t+(h/2),x+(h*k1/2)))
        k3=f(t+(h/2),x+(h*k2/2))
        k4=f(t+h,x+(h*k3))
        t=t+h 
        x= x + (h/6)*(k1+(2(k2+k3))+k4) 
        y=f(t,x) 
    end
    plot(X,Y) 
end
```

```julia
f(t,x)=-(cos(0.866*t)+(1.732*sin(0.866*t)))/exp(t/2)
```

```julia
Runge(f,1,0,0,0.001,100)
```

```julia
function Euler(f,x0,y0,t0,h,N) 
    T = [] 
    X = []
    Y=[]
    t=t0 
    x = x0
    y=y0
    while t<=N 
        push!(T,t) 
        push!(X,x) 
        push!(Y,y)
        t=t+h
        s=f(t,x)
        x = x + h*s 
        y=f(t,x)
    end  
    plot(X,Y,"black") 
end
```

```julia
function Runge4(f,x0,h,N) 
 for i in 1:5 
    T = []
    X = []
    a=rand(linspace(1,x0,101))
    println(a)
    t = (1/a)+h
    x = -1/t 
    while t<=N
        push!(T,t)
        push!(X,x)
        k1=(f(t,x,a)) 
        k2=(f(t+(h/2),x+(h*k1/2),a))
        k3=f(t+(h/2),x+(h*k2/2),a)
        k4=f(t+h,x+(h*k3),a)
        t=t+h
        x= x + (h/6)*(k1+(2(k2+k3))+k4)
    end
    plot(T,X)
    end
end
```

```julia
function f4(t,x,a)
    return (a^2)/((1-(a*t))^2)
end
```

```julia
function RungeM(f,x10,x20,t0,h,N,A) 
    a=A[1,1]
    b=A[1,2]
    c=A[2,1]
    d=A[2,2]
    T = []
    X1 = []
    X2=[]
    Y1 = []
    Y2=[]
    t = t0 
    x1 = x10
    x2 = x20
    y1 = (a*x1)+(b*x2)
    y2=(c*x1)+(d*x2)
    while t<=N
        push!(T,t)
        push!(X1,x1)
        push!(X2,x2)
        push!(Y1,y1)
        push!(Y2,y2)
        k1=(f(t,x1,x2,a,b))
        k2=(f(t+(h/2),x1+(h*k1/2),x2+(h*k1/2),a,b))
        k3=f(t+(h/2),x1+(h*k2/2),x2+(h*k2/2),a,b)
        k4=f(t+h,x1+(h*k3),x2+(h*k3),a,b)
        l1=(f(t,x1,x2,c,d))
        l2=(f(t+(h/2),x1+(h*l1/2),x2+(h*l1/2),c,d))
        l3=f(t+(h/2),x1+(h*l2/2),x2+(h*l2/2),c,d)
        l4=f(t+h,x1+(h*l3),x2+(h*l3),c,d)
        t=t+h
        x1= x1 + (h/6)*(k1+(2(k2+k3))+k4)
        x2= x2 + (h/6)*(l1+(2(l2+l3))+l4)
        y1=f(t,x1,x2,a,b)
        y2=f(t,x1,x2,c,d)
    end
    plot(X1,Y1)
    plot(X2,Y2)
end
```

```julia
A1 = [1 0;1 0]
A2 = [-1 0;-1 0]
A3 = [1 -2;0 0]
A4 = [1 -0.5;0 0]
A5 = [0 1;-1 0]
A6 = [1 1;-1 1]
A7 = [-1 1;-1 -1]
A8 = [-1 -1;1 -1]
A9 = [1 4;-2 1]
A10 = [-1 -1;0 -1]
```
