```julia
function lya(x,n,r)                 
    l=0    
for i in 1:n                                 
    x = r*x*(1-x)                       
    l = l+log(abs(r*(1-2*x)))                 
    end
    return l/n                              
    end
```

```julia
a=linspace(0,1,1000)
b=linspace(1,5,1000)
r =[]
s =[]
for i in 1:length(b)
    push!(r,bif.(a,b[i],1000))
    push!(s,ones(length(a))*b[i])
    end
```

```julia
using PyPlot
plot(s,r,"o",markersize=0.05)
plot(b,lya.(0.5,1000,b),"o",markersize=0.5)
title("Bifurcación")
title("Lyapunov")
grid()
```
