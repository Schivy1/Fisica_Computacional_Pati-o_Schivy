```julia
using PyPlot
```

```julia
function RK4(f,x0,t0,tf,h)
    t=t0:h:tf
    x=[x0[1]] # Posici√≥n
    v=[x0[2]]   #Velocidad
    for i in 2:length(t)
       #Pasos de R.k.4 
    k1=f(x0,t[i-1])
    k2=f(x0+(h/2)*k1,t[i-1]+h/2) 
    k3=f(x0+(h/2)*k2,t[i-1]+h/2) 
    k4=f(x0+h*k3,t[i]+h)
        
    x0=x0+(h/6)*(k1+2k2+2k3+k4)
        
        push!(x,x0[1])
        push!(v,x0[2])
    end
    return t,x,v
end
```

```julia
f(x,t)=[x[2]-(x[2])^3, -x[1]-(x[2])^2]
```

```julia
function espaciofase(g)  
for i in 1:100 
    x0=(rand(2)-rand(2))*6   
    random=RK4(g,x0,0.0,20.0,0.01)  
    plot(random[2],random[3])
    xlim(-5,5),ylim(-5,5)
    title("Espacio Fase")
         grid("on")
end
end
```

```julia
espaciofase(f)
```

Calcular el jacobiano de $\vec F = (y-y^3,-x-y^2)^{\text T}$. Sabemos que el jacobiano se ve como: $$J_F =\left(\begin{array}{cc} \frac{df_1}{dx} & \frac{df_1}{dy}  \\ \frac{df_2}{dx}  & \frac{df_2}{dy}  \end{array}\right)$$ con $ f_1=y-y^3$ y $f_2=-x-y^2$.

Entonces: $$J_F =\left(\begin{array}{cc} 0 & 1-3y^2 \\ -1 & -2y \end{array}\right)$$ En $\vec 0$ tenemos: $$J_{F(0,0)} =\left(\begin{array}{cc} 0 & 1 \\ -1 & 0 \end{array}\right)$$

Su determinante $\Delta = 1$ y su traza $\tau = 0$

## Ejercicio 4

```julia
k(x,t)=[x[1]+x[2]-x[1]*(x[1]^2+x[2]^2),-x[1]+x[2]-x[2]*(x[1]^2+x[2]^2)]
```

```julia
function espaciofase3(g)  
for i in 1:1000
    x0=3*(rand(2,1))-1   
    random=RK4(g,x0,-3,3,0.01)  
    plot(random[2],random[3])
    xlim(-2,2),ylim(-2,2)
    title("Espacio Fase")
        grid("on")
end
end
```

```julia
espaciofase3(k)
```

Viendo el jacobiano tenemos que: $$J_F =\left(\begin{array}{cc} 1-3x^2-y^2 & 1-2xy \\ -1-2xy & 1-3y^2-x^2 \end{array}\right)$$ En $\vec 0$ tenemos: $$J_{F(0,0)} =\left(\begin{array}{cc} 1 & 1 \\ -1 & 1 \end{array}\right)$$
Como la traza y determinante son mayores a $0$ vemos que en $\vec 0$ la funci√≥n tiene un puto fijo repulsor.
