```julia
using PyPlot
```

```julia
function Euler(f,x0,t0,tf,h)
    r=[]
    s=[]
    x = x0
    t = t0
    while t<=tf
        push!(r,x)
        push!(s,t)
    t = t + h
    x = x + h*f(x,t)
    end
    te = linspace(0,8,8/h+1)
    xe = map(y -> exp(-2.5*y), te) 
        scatter(s,r,color="green")
        plot(te,xe,color="blue")
end
```

```julia
f(x,y)=-2.5*x
```

```julia
Euler(f,1,0,8,1)
```

```julia
Euler(f,1,0,8,0.5)
```

```julia
Euler(f,1,0,8,0.2)
```

```julia
Euler(f,1,0,8,0.01)
```

```julia
function Euler2(f,x0,t0,tf,h1,h2,h3)
x3=x2=x1=zeros(1,tf)
t3=t2=t1=zeros(1,tf)
x1[1]=x2[1]=x3[1]=x0
t1[1]=t2[1]=t3[1]=t0
    for i in 2:tf
    t1[i] = t0 + (i-1)*h1
    x1[i] = x1[i-1] + h1*f(x1[i-1],t1[i-1])
    t2[i] = t0 + (i-1)*h2
    x2[i] = x2[i-1] + h2*f(x2[i-1],t2[i-1])
    t3[i] = t0 + (i-1)*h3
    x3[i] = x3[i-1] + h3*f(x3[i-1],t3[i-1])
    end
    te = linspace(0,8,8/h1+1)
    xe = map(y -> exp(-2.5*y), te) 
        scatter(t1,x1,color="red")
        scatter(t2,x2,color="green")
        scatter(t3,x3,color="yellow")
        plot(te,xe,color="blue")
end
```

```julia
function Euler2(f,x0,t0,tf,h1,h2,h3)
    r1=[]
    r2=[]
    r3=[]
    s1=[]
    s2=[]
    s3=[]
    x1 =x2 =x3 = x0
    t1=t2=t3 = t0
    while t1<=tf
        push!(r1,x1)
        push!(s1,t1)
    t1 = t1 + h1
    x1 = x1 + h1*f(x1,t1)
    end
    while t2<=tf
        push!(r2,x2)
        push!(s2,t2)
    t2 = t2 + h2
    x2 = x2 + h2*f(x2,t2)
    end
    while t3<=tf
        push!(r3,x3)
        push!(s3,t3)
    t3 = t3 + h3
    x3 = x3 + h3*f(x3,t3)
    end
    te = linspace(0,8,8/h1+1)
    xe = map(y -> exp(-2.5*y), te) 
        scatter(s1,r1,color="green")
        scatter(s2,r2,color="red")
        scatter(s3,r3,color="yellow")
        plot(te,xe,color="blue")
end
```

## Ejercicio 2

El ocilador armónico.

Sea $w=\sqrt{k}$:

$x=Ce^{wit}+Ae^{-wit}=(A+C)cos(wt)$

Derivando:

$\dot{x}=-w(A+C)sen(wt)$

$\ddot{x}=-w^2(A+C)cos(wt)=-k(A+C)cos(wt)=-kx$

Para las condiciones iniciales, $x(0)=\alpha=A+C$ y para toda $x'(0)=0$ porque sen(0)=0.

## Ejercicio 4

```julia
function EulerVec(x0,t0,k,h,N)
    r=[]
    s=[]
    m=[]
    x=x0 
    t=t0 
    w=sqrt(k) 
    a=-(w*sin(w*t0)) 
    for i in 1:N
        push!(r,x) 
        push!(s,t)
        push!(m,a) 
        u=-(w*sin(w*t))
        x=x+(h*u)
        a=a-(h*(k*cos(w*t))) 
        t=t+h
    end
    
    scatter(s,r,color="blue")
end
```

```julia
EulerVec(0,0,1,0.01,10000)
```

```julia
function Euler3(f,x0,t0,tf,h)
    r=[]
    s=[]
    x = x0
    t = t0
    while t<=tf
        push!(r,x)
        push!(s,t)
    t = t + h
    x = x + h*f(x,t)
    end
    te = linspace(0,8,8/h+1)
    xe = map(y -> 4 + 8*y - 5*(y*y), te) 
        scatter(s,r,color="green")
        plot(te,xe,color="blue")
end
```
