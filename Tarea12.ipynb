```julia
function eulerimp(f,x0,t0,tf,h)
    r = [] 
    s = []
    t=t0 
    x = x0 
    while t<tf
        push!(r,t)
        push!(s,x)
        t=t+h
        j=f(x,t)
        x = x + h*j
        
    end  
        scatter(r,s,color="green",label="Solución método de Euler", s = 10)
    legend()
    grid()
end
```

```julia
function puntoMedio(f,x0,t0,h,N)
s=[]
r=[]
x = x0 
t = t0
        while t<N
        push!(s,x)
        push!(r,t)
        t =t + h
        t2 = t + h/2
        x = x +h*f(x+(h/2)*f(x,t),t2)
    end
    te = linspace(0,8,8/h+1)
    xe = map(y -> exp(-2.5*y), te) 
        scatter(r,s,color="green",label="Solución método de Euler", s = 10)
    legend()
        plot(te,xe,color="blue",label="Solución exacta")
    legend()
    grid()
end
```

## Ejercicio 3

```julia
function RungeKutta(f,x0,t0,N,h)
s=[]
r=[]
x = x0 
t = t0
    while t<=N
        push!(s,x)
        push!(r,t)
        k1 = f(t,x)
        k2 = f(t+h/2,x+(k1*h)/2)
        k3 = f(t+h/2,x+(k2*h)/2)
        k4 = f(t+h,x+k3*h)
        t = t + h
        x = x + (h/6)*(k1+2k2+2k3+k4)
    end
        scatter(r,s,color="green",label="Solución por Runge Kutta", s = 10)
    legend()
    grid()
end
```

## Ejercicio 6

```julia
function RungeKutta2D(f,x0,v0,t0,tf,h)
    T=[]
    X=[]
    V=[]
    x=x0
    v=v0
    t=t0
    while t<tf
        push!(X,x)
        push!(V,v)
        push!(T,t)
        l1 = h*f(x,v,t)
        k1 = h*v
        l2 = h*f(x+(k1/2),v+(l1/2),t+(h/2))
        k2 = h*(v+(l1/2))
        l3 = h*f(x+(k2/2),v+(l2/2),t+(h/2))
        k3 = h*(v+(l2/2))
        l4 = h*f(x+k3,v+l3,t+h)
        k4 = h*(v+l3)
        t = t + h
        v = v + ((l1+(2*l2)+(2*l3)+l4)/6)
        x = x + ((k1+(2*k2)+(2*k2)+k2)/6)
    end
        subplot(2,1,1)
    plot(T, X, color = "blue")
    legend()
    title("Gráfica de la función")
    grid()
    subplot(2,1,2)
    plot(T,V, color="red")
    legend()
    title("Gráfica de la derivada de la función")
    grid() 
end
```

```julia
RungeKutta2D((x,v,t)->-2x,1,0,0,10,0.1)
```

```julia
function Euler2D(f,f2,x0,v0,t0,tf,h)
X = []
V = []
T = []
x = x0
v = v0
t = t0
    while t<tf
        push!(V,v)
        push!(X,x)
        push!(T,t)
        v = v + f(t,v,x*h)
        x = x + f2(t,v,x*h)
        t = t + h
    end
    figure()
    subplot(2,1,1)
    plot(T, X, color = "blue")
    legend()
    title("Gráfica de la función")
    grid()
    subplot(2,1,2)
    plot(T,V, color="red")
    legend()
    title("Gráfica de la derivada de la función")
    grid()
end 
```

```julia
function Eulerenergia(f,f2,x0,v0,t0,tf,h,m,r)
X = []
V = []
T = []
K = []
U = []
Et = []
x = x0
v = v0
t = t0
u = 0.5*r*(x0*x0)
k = 0
et = k +u
    while t<tf
        push!(V,v)
        push!(X,x)
        push!(T,t)
        push!(U,u)
        push!(K,k)
        push!(Et,et)
        v = v + f(t,v,x*h)
        x = x + f2(t,v,x*h)
        t = t + h
        u = 0.5*r*(x*x)
        k = 0.5*m*(v*v)
        et = k+u
    end
    figure()
    subplot(2,1,1)
    plot(T, K, color = "blue")
    plot(T, U, color = "green")
    legend()
    title("Energía cinética y potencial")
    grid()
    subplot(2,1,2)
    plot(T,Et, color="red")
    legend()
    title("Energía total")
    grid()
end 
```
